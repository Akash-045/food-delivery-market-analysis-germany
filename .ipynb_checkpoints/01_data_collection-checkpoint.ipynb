```python
import pandas as pd
import time
from google_play_scraper import reviews, Sort, app
from datetime import datetime
import os
os.makedirs('data', exist_ok=True)
```

```python
## Defined apps

apps = {
    'Wolt': 'com.wolt.android',
    'Lieferando': 'com.yopeso.lieferando',
    'Uber Eats': 'com.ubercab.eats'
}
```

```python
## Scraping reviews

all_reviews = []

for app_name, app_id in apps.items():
    print(f"Scraping {app_name}...")
    result, _ = reviews(
        app_id,
        lang='de',
        country='de',
        sort=Sort.NEWEST,
        count=2000
    )
    for r in result:
        r['app_name'] = app_name
    all_reviews.extend(result)
    time.sleep(3)
    
print(f"Total reviews collected: {len(all_reviews)}")
```

```python
## Coverted to dataframe and saved

df = pd.DataFrame(all_reviews)
df['scraped_at'] = datetime.now()
df.to_csv('data/raw_reviews.csv', index=False)
print(df.shape)
df.head()
```
